```python
#List of modules used
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

#Grading for Calculus 2 Summer Term
#There has been 3 exams in this semester.
#Two midterms and a final exam.
#Each midterm will effect total score %20.
#Final exam will effect the results %60.

class Student:

  def __init__(self, name, surname, midterm_1 = 0, midterm_2 = 0, final = 0):
    self.name = name
    self.surname = surname
    self.midterm_1 = float(midterm_1)
    self.midterm_2 = float(midterm_2)
    self.final = float(final)


  #Calculates the total score. Midterms %20, Final %60.
  def calculateFinalScore(self):
    result =  self.midterm_1 * 0.2 + self.midterm_2 * 0.2 + self.final * 0.6
    return result

  #Calculates the final grade of a student purely based on their scores.
  def calculateGrade(self):

    
      total = self.calculateFinalScore()
      if 100 >= total >= 95:  
        return 'A1 Passed'
      elif 95 > total >= 90:  
        return 'A2 Passed'
      elif 90 > total >= 85:  
        return 'A3 Passed'
      elif 85 > total >= 80:  
        return 'B1 Passed'
      elif 80 > total >= 75:  
        return 'B2 Passed'
      elif 75 > total >= 70:  
        return 'B3 Passed'
      elif 70 > total >= 65:  
        return 'C1 Passed'
      elif 65 > total >= 60:
        return 'C2 Passed'
      elif 60 > total >= 55:  
        return 'C3 Passed'
      elif 55 > total >= 50:  
        return 'D Passed'
      if 50 > total >= 0:  
        return 'F3 Failed'

      else:
        return None


  #Gives information about the student. Can be called as Student.studentInfo()
  def studentInfo(self):
    description = f""" 
                      Student Name:{self.name}
                    Student Surname:{self.surname}
                          Midterm 1:{self.midterm_1}
                          Midterm 2:{self.midterm_2}
                              Final:{self.final}
                                    """
    
    return description

class Classroom(Student):

  def __init__(self, student_list = []):
     self.student_list = student_list

  def classInfo(self):
    for student in self.student_list:
      print(student.studentInfo())
    return None

    #This method calls Pandas DataFrame to_excel function.
    #Directly saves your results to your current location.
  def saveResults(self, save_name = 'results.xlsx'):
    self.gradeClass().to_excel(save_name)
    return self.gradeClass()

  #Classroom operations can be added.
  def addStudent(self, student_info):
    self.student_list.append(student_info)
    return self.student_list

  def removeStudent(self, student_id):
    self.student_list.pop(student_id - 1)
    return self.student_list
      
  #Use this if you want to grade the entire class and save it as Pandas DataFrame.
  #can be called as results_df = Classroom.gradeClass()
  #for Gaussian grading (or bell curve grading) write as gradeClass(bell = True)

  def gradeClass(self, bell = False):

    student_dict = dict()
    id = 1

    for student in self.student_list:
      info_std = [student.name, student.surname, student.midterm_1, student.midterm_2, student.final, student.calculateFinalScore(), student.calculateGrade()]
      student_dict[id] = info_std
      id += 1

      
    student_df = pd.DataFrame.from_dict(student_dict, orient = 'index', columns = ['Name', 'Surname', 'Midterm_1', 'Midterm_2', 'Final', 'Total_Score', 'Grade & Results'])

    return student_df

  

```
